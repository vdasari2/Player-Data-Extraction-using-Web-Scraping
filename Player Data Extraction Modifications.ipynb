```python
toppts = df.loc[: ,['Player Name','PTS']].sort_values('PTS', ascending=False).iloc[0:6]
toppts
topc_pts = df.loc[: ,['Player Name','C_PTS']].sort_values('C_PTS', ascending=False).iloc[0:6]
topc_pts
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1)
ax1.plot(toppts['Player Name'],toppts['PTS'], linewidth=2.5, color = 'red', linestyle='--')
ax2.plot(topc_pts['Player Name'],topc_pts['C_PTS'], linewidth=2.5, color = 'red', linestyle='--')
plt.subplots_adjust(top=1.5, bottom=0.1, left=0.1, right=0.9, hspace=2, wspace=2)
plt.xticks(fontsize = 10)
plt.tight_layout()
plt.xlabel("Points")
plt.ylabel("Teams")
ax1.set_title('Player vs PTS') 
ax2.set_title('Player vs C_PTS')
plt.show()
```
